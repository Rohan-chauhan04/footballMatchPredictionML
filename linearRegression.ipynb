```python
def create_classification_model(settings: ml_edu.experiment.ExperimentSettings, metrics):
    # Inputs
    inputs = {name: keras.Input(shape=(1,), name=name) for name in settings.input_features}
    concatenated_inputs = keras.layers.Concatenate()(list(inputs.values()))

    # Output: 3 classes (H, D, A)
    outputs = keras.layers.Dense(units=3, activation='softmax')(concatenated_inputs)

    model = keras.Model(inputs=inputs, outputs=outputs)

    model.compile(
        optimizer=keras.optimizers.RMSprop(learning_rate=settings.learning_rate),
        loss='categorical_crossentropy',
        metrics=metrics
    )

    return model


def train_model(
    experiment_name: str,
    model: keras.Model,
    dataset: pd.DataFrame,
    label_name: str,
    settings: ml_edu.experiment.ExperimentSettings,
) -> ml_edu.experiment.Experiment:
  """Train the model by feeding it data."""

  # Feed the model the feature and the label.
  # The model will train for the specified number of epochs.
  features = {name: dataset[name].values for name in settings.input_features}
  label = dataset[label_name].values
  history = model.fit(x=features,
                      y=label,
                      batch_size=settings.batch_size,
                      epochs=settings.number_epochs)

  return ml_edu.experiment.Experiment(
      name=experiment_name,
      settings=settings,
      model=model,
      epochs=history.epoch,
      metrics_history=pd.DataFrame(history.history),
  )

print("SUCCESS: defining linear regression functions complete.")
```

```python
import pandas as pd
from tensorflow import keras
import ml_edu

# 1️⃣ Load your dataset
url = "https://raw.githubusercontent.com/tarekmasryo/football-matches-2025-data/main/data/football_matches_2024_2025.csv"
football_df = pd.read_csv(url)

# Input
X = football_df[['home_team_id', 'away_team_id']].astype(int)
X_encoded = pd.get_dummies(X, columns=['home_team_id', 'away_team_id'])

# Target
y_encoded = pd.get_dummies(football_df['match_outcome'])

# Settings
input_features = X_encoded.columns.tolist()
settings_1 = ml_edu.experiment.ExperimentSettings(
    learning_rate = 0.001,
    number_epochs = 20,
    batch_size = 50,
    input_features = input_features
)

metrics = [keras.metrics.CategoricalAccuracy(name='accuracy')]
model_1 = create_classification_model(settings_1, metrics)


# Combine X_encoded and y_encoded for train_model
train_df = X_encoded.join(y_encoded)
experiment_1 = train_model('match_result', model_1, train_df, y_encoded.columns.tolist(), settings_1)

```

```python
import pandas as pd

# New match
new_match = pd.DataFrame({'home_team_id': [66], 'away_team_id': [62]})

# One-hot encode to match training columns
new_match_encoded = pd.get_dummies(new_match, columns=['home_team_id', 'away_team_id']).astype(float)

# Ensure all columns match training data and are in the correct order efficiently
new_match_encoded = new_match_encoded.reindex(columns=X_encoded.columns, fill_value=0.0)
```

```python
# Prepare new match input
prediction_input = {col: new_match_encoded[col].values for col in new_match_encoded.columns}

# Predict probabilities
pred_probs = model_1.predict(prediction_input)

# Pick highest probability class
pred_class_index = pred_probs.argmax(axis=1)[0]
pred_class = y_encoded.columns[pred_class_index]

print("Predicted match outcome:", pred_class)

```
